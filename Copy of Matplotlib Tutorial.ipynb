#MatplotLib Tutorial
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Streamlit, wxPython, Qt, or GTK+.

#Some of the major Pros of Matplotlib are:

Generally easy to get started for simple plots
Support for custom labels and texts
Great control of every element in a figure
High-quality output in many formats
Very customizable in general

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 2 * np.pi, .1)
y = np.cos(x)
plt.title("sine wave form")

# Plot the points using matplotlib
plt.plot(x, y)
plt.show()
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)


plt. subplot(2, 1,1)

plt.plot(x, y_sin,'r--')
plt.title('Sine')


plt.subplot(2, 1, 2)
plt.plot(x, y_cos,'g--')
plt.title('Cosine')

plt.show()
```

```python
## Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.subplot(1,3,1)
plt.bar(x, y)

plt.subplot(1,3,2)
plt.plot(x, y, color = 'g')
plt.subplot(1,3,3)
plt.bar(x2, y2, color = 'r')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

#Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0,.1, 0)  # explode 1st slice

# Plot
plt.pie(sizes,labels=labels,colors=colors,explode=explode,autopct='%1.2f%%')
plt.axis('equal')
plt.show()
```
