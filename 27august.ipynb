```python
#constructor overriding
class parent():
  def __init__(self):
    print("this is parent class")
class child(parent):
  def __init__(self):
    print("this is child class")
    super().__init__()
```

```python
p=parent()
c=child()
```

```python
#method overriding
class parent():
  def show(self):
    print("this is parent class")
class child(parent):
  def show(self):
    print("this is child class")
    super().show()
```

```python
a=parent()
b=child()
```

```python
from abc import ABC,abstractmethod
class Defence(ABC):
  @abstractmethod
  def area(self):
    pass
  @abstractmethod
  def weapons(self):
    pass
```

```python
class army(Defence):
  def area(self):
    print("Land")
  def weapons(self):
    print("ak47")
```

```python
class navy(Defence):
  def area(self):
    print("water")
  def weapons(self):
    print("cannon")
```

```python
class airforce(Defence):
  def area(self):
    print("air")
  def weapons(self):
    print("missile")
```

```python
a=army()
n=navy()
b=airforce()
```

```python
from abc import ABC, abstractmethod
class Defence(ABC):
  @abstractmethod
  def area(self):
    pass
  def guns(self,obj):
    if obj==a:
      print("a","b",'c')
    elif obj==n:
      print("d",'e','f')
```

```python
class employee:
  def __init__(self,name):
    self.name=name
    self._salary=30000
  def display(self):
    print(self.name)
    print(self._salary)
```

```python
a=employee("ram")
```

```python
a.display()
```

```python
print(a._salary)
```

```python
class finance:
  def __init__(self):
    self.__revenue=100000 #private variable
    self.no_of_sales=1323
  def display(self):
    print(f"revenue is : {self.__revenue} and no of sales is {self.no_of_sales}")
    self.__revenue=200000
    print(f"revenue is : {self.__revenue} and no of sales is {self.no_of_sales}")
```
