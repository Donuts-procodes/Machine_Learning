```python
n=int(input("enter the no.\t"))
s=n
b=len(str(n))
sum1=0
while n!=0:
  r=n%10
  sum1=sum1+(r**b)
  n=n//10
if s==sum1:
  print("armstrong")
else:
  print("not armstrong")
```

```python
a=int(input("enter any no\t"))
n=str(a)
b=len(n)
sum=0
i=0
while i<b:
  x=int(n[i])
  sum=sum+x**3
  i=i+1
if sum==a:
  print("armstrong")
else:
  print("not armstrong")
```
