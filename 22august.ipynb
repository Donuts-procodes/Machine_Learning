```python
#multiple inheritence
class father():
  def __init__(self):
    print("father constructor")
  def show(self):
    print("father show")
```

```python
class mother():
  def __init__(self):
    print("mother constructor")
  def print(self):
    print("mother print")
```

```python
class son(father,mother):
  def __init__(self):
    print("son constructor")
    super().__init__()
  def display(self):
    print("son display")
```

```python
x=son()
x.display()
x.show()
x.print()
```

```python
class D(B):
  def __init__(self):
    print("D constructor")
    super().__init__()
```

```python
class duck():
  def talk(self):
    print("quack quack")
```

```python
class human():
  def talk(self):
    print("hello")
```

```python
d=duck()
h=human()
```

```python
d.talk()
h.talk()
```

```python
def call(o):
  o.talk()
call(human)
```

```python
class vishesh():
  def study(self):
    print("python")
class piyush():
  def study(self):
    print("java")
class amit():
  def study(self):
    print("c++")
class aman():
  def training(self):
    print("c")
```

```python
v=vishesh()
p=piyush()
a=amit()
b=aman()
```

```python
a.study()
```

```python
def call(o):
  if hasattr(o,'study'):
    o.study()
  elif hasattr(o,'training'):
    o.training()
```

```python
call(p)
```

```python
call(a)
```

```python
call(v)
```

```python
call(b)
```
