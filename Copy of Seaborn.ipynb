# Here's what you need to know about the Iris dataset

## Contents:
The dataset contains measurements of 150 iris flowers, equally divided into 3 species: Setosa, Versicolor, and Virginica.

## Features:
For each flower, 4 features are recorded: sepal length, sepal width, petal length, and petal width. These are all numerical values measured in centimeters.

## Target Variable:
The goal is typically to predict the species of an iris flower based on its features. This is a classification task with 3 possible classes (the 3 species).
