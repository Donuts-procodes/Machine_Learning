```python
#recursive
def fact(n):
  if n==0:
    return 1
  else:
    print(n)
    return n*fact(n-1)
```

```python
fact(5)
```

```python
def abc(n):
  if n==0:
    return 0
  else:
    return n+abc(n-1)
```

```python
abc(10)
```

```python
def mult(a,b):
  if b==1:
    print(b)
    return a
  else:
    print(b)
    return a+mult(a,b-1)
```

```python
mult(2,10)
```
