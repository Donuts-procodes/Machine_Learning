```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("heart.xls")
```

```python
df
```

```python
x=df.iloc[:0:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
x_scaled=st.fit_transform(x)
```

```python
x_scaled
```

```python
from sklearn.model_selection import train_test_split
y = y[:1024]  # Trim y to match x_scaled
```

```python

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.25, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
y_pred_test=lg.predict(x_test)
```

```python
y_pred_test
```

```python
y_pred_train=lg.predict(x_train)
y_pred_train
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score,f1_score
```

```python
lg.predict_proba(x_train)
```

```python
accuracy_score(y_train,y_pred_train)
```

```python
accuracy_score(y_test,y_pred_test)
```

```python
pd.DataFrame(confusion_matrix(y_test,y_pred_test),columns=list(range(0,2)))
```

```python
import seaborn as sns
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred_test),annot=True)
```
