```python
class circle:
  def __init__(self,radius):
    self.radius=radius
  def area(self):
    return 3.14*self.radius*self.radius
  def perimeter(self):
    return 2*3.14*self.radius
```

```python
a=circle(7)
a.area()
a.perimeter()
```
