```python
class car():
  @staticmethod
  def show(x):
    z=x**2
    return z
```

```python
a=car()
a.show(10)
```

```python
class sam():
  @staticmethod
  def show(m,p):
    d=p*.2
    d1=p-d
    print("model number is :",m)
    print("price is :",d1)
```

```python
x=sam()
```

```python
x.show("abc",10000)
```

```python
class employee():
  def __init__(self,name,salary,age):
    self.name=name
    self.salary=salary
    self.age=age
  def show(self):
    print("name is :",self.name)
    print("age is :",self.age)
    print("salary is :",self.salary)
    return self.salary

  a=employee('abc',50000,23)
  new=a.show()

  class increment():
    @staticmethod
    def inc(x):
      y=x+10000
      print("new salary is :",y)

  m=increment()
  m.inc(new)
```

```python
class father():
  def __init__(self):
    self.prop=10000000
  def show(self):
    print(f"father class {self.prop}")
```
