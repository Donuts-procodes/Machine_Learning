```python
def order():
    a = input("Do you want to order? (yes or no):\t")
    if a == 'yes':
        menu = {"tea": 100, "coffee": 150, "snacks": 200}
        print("Menu:", menu)
        return a, menu
    else:
        print("Thank you!")
        return a, {}

def decor(f):
    def inner():
        sum = 0
        a, menu = f()  # Call the original function to get the user's decision and menu
        if a == 'yes':
            while True:
                print(menu)
                c = input("Enter your choice: ")
                if c in menu:
                    sum += menu[c]
                    print(f"Your total is: {sum}")
                else:
                    print("Invalid choice, please try again.")

                # Ask if they want to order something else
                new = input("Would you like anything else from the menu? (yes or no):\t")
                if new.lower() != 'yes':  # Stop if they don't want anything else
                    break
            print(f"Your final total is: {sum}")
        else:
            print("Thank you!")
    return inner

@decor
def order_decorated():
    return order()

# Use the decorated function
order_decorated()
```
