```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv("dataset.csv")
```

```python
df
```

# decision tree

```python
x=df.drop(columns=['target'])
y=df['target']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=2)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_pred_train=dt.predict(x_train)
y_pred_test=dt.predict(x_test)
```

```python
print("accuracy score of train: ",accuracy_score(y_train,y_pred_train))
print("accuracy score of test: ",accuracy_score(y_test,y_pred_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'max_depth': [3, 4, 5, 6, 7, 8, 9, 10],
    'min_samples_leaf': [1, 2, 3, 4, 5, 6, 7, 8],
    'min_samples_split': [2, 3, 4, 5, 6, 7, 8],
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'random_state': [0, 42]
}

dt=DecisionTreeClassifier()
# dt.fit(x_train,y_train)

# Perform Grid Search
grid_search = GridSearchCV(dt, param_grid, cv=5, n_jobs=-1,verbose=1)
grid_search.fit(x_train, y_train)

# Print best parameters
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
b=grid_search.best_params_
b    
```

```python
dtree2=DecisionTreeClassifier(**b)
dtree2.fit(x_train,y_train)
```

```python
y_pred_train2=dtree2.predict(x_train)
y_pred_test2=dtree2.predict(x_test)
print("accuracy score of train: ",accuracy_score(y_train,y_pred_train2))
print("accuracy score of test: ",accuracy_score(y_test,y_pred_test2))
```
