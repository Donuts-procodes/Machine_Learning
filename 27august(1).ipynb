```python
class Book1():
  def __init__(self,pages):
    self.pages=pages
```

```python
class Book2():
  def __init__(self,pages):
    self.pages=pages
  def __add__(self,other):
    return self.pages+other.pages
```

```python
b1=Book1(100)
b2=Book2(200)
```

```python
b2+b1
```

```python
class a():
  def __init__(self,num):
    self.num=num
class b():
  def __init__(self,num):
    self.num=num
  def __lt__(self,other):
    return self.num<other.num
```

```python
a=a(12)
b=b(11)
```

```python
b<a
```

```python
class calculate():
  def sum(self,a,b):
    print(a+b)
  def sum(self,a,b,c):
    print(a+b+c)
```

```python
c=calculate()
```

```python
class abc():
  def aaa(self,a,b,c):
    d=a+b+c
    return d
```

```python
c=abc()
```

```python
c.aaa(10,20,10)
```

```python
class Mm:
  def sum(self,a=None,b=None,c=None):
    if a!=None and b==None and c==None:
      c=a*a
    elif a!=None and b!=None and c==None:
      c=a*b
    else:
      c=a+b+c
    return c
```

```python
s=Mm()
```

```python
s.sum(10)
```

```python
class father():
  def __init__(self):
    print("father")
  def x(self):
    print("dad")
class son(father):
  def __init__(self):
    print("son")
  def x(self):
    #print("child")
    super().x()
```

```python
s=son()
f=father()
```

```python
s.x()
```

```python
class father:
  def result(self,a,b):
    print("addition:",a+b)
class son(father):
  def result(self,a,b):
    super().result(a,b)
    print("multiplication:",a*b)
```

```python
s=son()
```

```python
s.result(20,30)
```

```python
class square():
    def area(self, x):
        print("area of square:", x**2)

class circle(square):
    def area(self, x):
        super().area(x)
        print("area of circle:", 3.14 * (x**2))
shape = circle()
shape.area(5)
```
