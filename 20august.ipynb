```python
class car():
  def __init__(self):
    self.model="x1"
    self.year=2019
    self.prize=1000000
    self.f_type='petrol'

  def discount(self):
    x=int(input("enter your dicount:"))
    self.prize=self.prize-self.prize*(x/100)
    d=self.prize
    print(f"after discount is {d}")

c=car()
c.discount()
```

```python
print(c.model)
print(c.year)
print(c.prize)
```

```python
class student():
  def __init__(self,n,age,m):
    self.name=n
    self.age=age
    self.marks=m
  def eligiblity(self):
    if self.age>18:
      print("eligible")
    else:
      print("not eligible")
  def scholarship(self):
    if self.marks>450:
      print("eligible for scholarship")
    else:
      print("not eligible for scholarship")
```

```python
a=student('abc',18,459)
```

```python
a.eligiblity()
```

```python
a.scholarship()
```

```python
'''
types of variables in classes;
there are two types:
1.instants variables:work according to the object only and work only for the instance only
2.class/static variables:can be modified anywher in a class
'''
```

```python
class sample():
  def __init__(self):
    self.x=10 #instance varibale
  def modify(self):
    self.x=self.x+1
    print(self.x)
```

```python
s1=sample()
print(s1.x)
s1.modify()
s1.modify()
```

```python
s2=sample()
s2.modify()
```

```python
class variable():
  x=10#class variable
  @classmethod #decorator
  def modify(cls):#class method
    cls.x=cls.x+1
    print(cls.x)
```

```python
a=variable()
print(a.x)
a.modify()
```
